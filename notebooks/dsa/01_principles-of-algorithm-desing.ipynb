# Principles of Algorithm Design

I have defined algorithms in the previous section, but let's have another definition of it.
Algorithms are a sequential set of instructions to execute a program for a particular task.

Problem solving(in dsa context) is formed with two main things:

- <mark>Efficient</mark> mechanism to store, retrieve, and manage data.
- <mark>Efficient</mark> algorithm with a finite set of instructions to solve the problem.

So, what are the characteristics of an efficient algorithm:

- Specific.
- Clear.
- Every instruction should be executable in finite amount of time.
- Clear input and output.

## Algorithm Paradigms

There are three main algorithm paradigms, or in other ways, there are three main approaches for algorithm design.
Divide and conquer, dynamic programming, and greedy algorithms.

I will probably write detailed pages about each one of them so yeah, lets move on.

> Recursion and backtracking are popular and useful techniques in divide and conquer problems.

### Recursion

It is an algorithmic technique where a function solve the problem by calling a smaller, identical version
of itself till it reaches a simple and directly solvable problem.

- Base Case: Tell when the recursion should be terminated.
- Recursive Case: The function call itself.

### Backtracking

It is an algorithmic technique where it solve the problem by trying to build the solution incrementally, one piece at a time.

## Time Complexity

Time complexity in algorithms is a measure of how the runtime of an algorithm grows as the size of the input increase.
When analyzing the time complexity of an algorithm, you mainly focus on the operation that have the most significant effect.

Let's observe the time complexity of some famous operations:

- Iteration: O(n)
- Linear search: O(n)
- Retrieval in hashmaps and arrays(by index): O(1)
- Merge sort: O(nlogn)

There are three cases for analyzing the time and memory complexity of algorithms:

- Best Case(Ω - big omega notation): Yeah who give a f. No, just kidding. OK, calculating the best case complexity of an algorithm requires
analyzing the minimum steps or resources an algorithm need to be completed.
- Average Case(Θ - big theta notation): Estimating an algorithm's typical performance assuming input follows a specific probability

Benchmarking: Benchmarking involves comparing the performance of different algorithms(hardware, or basically works in this context) or implementations
by the same set of inputs.
